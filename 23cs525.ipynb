```python
import pandas as pd
df=pd.read_csv('/content/mtcars.csv')
print(df)
```

bar plot of am

```python
import pandas as pd
import matplotlib.pyplot as plt
auto_count=(df['am']==0).sum()
manul_count=(df['am']==1).sum()
labels=['automatic','manul']
counts=[auto_count,manul_count]
plt.bar(labels,counts,color=['red','blue'])
plt.title('bar of am')
plt.xlabel('values of am')
plt.ylabel('sum of am')
plt.grid(axis='y')
plt.show()
```

relationship btw wt and mpg


```python
import matplotlib.pyplot as plt
x=df['wt']
y=df['mpg']
plt.scatter(x,y,color='red')
plt.title('relationship btw wt and mpg')
plt.xlabel('wt')
plt.ylabel('mpg')
plt.grid(axis='y')
plt.show()
```
